## Задание 1

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером. Если да, то функция должна возвращать отдельно номер и регион.

```python
#Не учитываем прицепы, дипломатов, МВД, спецтехнику, советские номера и прочее, только легковые гражданские автомобили.
import re

def main():
    item = input('Введите транспортный номер, включая регион: ').lower()
    print(check_car_number(item))
    
def check_car_number(item):
    pattern = r'([авекмнорстух]{1}[\d]{3}(?<!000)[авекмнорстух]{2})([\d]{2,3})'
    result = re.match(pattern, ''.join(item.split()), flags = re.I)
    if result:
        return 'Номерной знак {0}, регион {1}'.format(result[1], result[2])
    else:
        return 'Номерной знак "{0}" не является валидным'.format(item)

main()
```

## Задание 3
При помощи регулярных выражений напишите функцию, которая будет возвращать акроним переданного в нее текста.

```python
import re

def main():
    item = input("Введите текст, из которого нужно извлечь акроним: ").strip()
    return (get_acronim(item))

def get_acronim(item):
    pattern = r"(?<=[\S])[\S]*\s*"
    result_set = re.split(pattern, item)
    return "Акроним следующий: {0}".format(''.join(result_set))

main()
```

## Задание 4
Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки.

```python
import re

def main():
    item = input("Введите текст, из которого нужно удалить повторы: ").strip()
    return (replace_duplicates(item))

def replace_duplicates(item):
    pattern = r"(.+)\1"
    replace_string = r"\1"
    while re.search(pattern, item):
        item = re.sub(pattern, replace_string, item, flags = re.I)
    return "Результирующий текст: {0}".format(item)

main()
```
